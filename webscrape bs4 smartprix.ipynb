#  webscrapping smartprix website

##  importing nesseccery Libraries

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

## reading HTML files mobile

```python
with open('smartprix_mobiles.html','r',encoding='utf-8') as f:
    html = f.read()
```

## creating soup Object

```python
soup = BeautifulSoup(html,'lxml')
container = soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'})
```

## Extacting Data of all Mobiles

```python
# Create lists to store data

name = []         # List to store product names
price = []        # List to store product prices
scores = []       # List to store product scores
ratings = []      # List to store product ratings
sim = []          # List to store SIM card information
processor = []    # List to store processor details
ram = []          # List to store RAM specifications
battery = []      # List to store battery details
display = []      # List to store display specifications
camera = []       # List to store camera specifications
card = []         # List to store memory card support information
os = []           # List to store operating system details

# Loop through the container to extract product details
for i in container:
    try:
        # Extract product name
        name.append(i.find('h2').text)
    except AttributeError:
        name.append(np.nan)
    
    try:
        # Extract product price
        price.append(i.find('span', {'class': 'price'}).text)
    except AttributeError:
        price.append(np.nan)

    # Extract product score
    try:
        score_element_1 = i.find('div', {'class': 'score rank-1-bg'})
        score_element_2 = i.find('div', {'class': 'score rank-2-bg'})
        score_element_3 = i.find('div', {'class': 'score rank-4-bg'})
        score_element_4 = i.find('div', {'class': 'score rank-5-bg'})
        score_element_5 = i.find('div', {'class': 'score rank-3-bg'})

        if score_element_1:
            score = score_element_1.find('b').text
        elif score_element_2:
            score = score_element_2.find('b').text
        elif score_element_3:
            score = score_element_3.find('b').text
        elif score_element_4:
            score = score_element_4.find('b').text
        elif score_element_5:
            score = score_element_5.find('b').text
        else:
            score = np.nan
        scores.append(score)
    except AttributeError:
        scores.append(np.nan)

    # Extract product rating
    try:
        rating_span = i.find('span', class_='sm-rating')
        if rating_span:
            rating = rating_span['style'].split(':')[1].strip(';')
        else:
            rating = np.nan
        ratings.append(rating)
    except (AttributeError, IndexError, TypeError):
        ratings.append(np.nan)
        
    # Extract other specifications
    x = i.find('ul',{'class':'sm-feat specs'}).find_all('li')
    try:
        sim.append(x[0].text)
    except:
        sim.append(np.nan)
    try:
        processor.append(x[1].text)
    except:
        processor.append(np.nan)
    try:    
        ram.append(x[2].text)
    except:
        ram.append(np.nan)
    try:
        battery.append(x[3].text)
    except:
        battery.append(np.nan)
    try:
        display.append(x[4].text)
    except:
        display.append(np.nan)
    try:
        camera.append(x[5].text)
    except:
        camera.append(np.nan)
    try:
        card.append(x[6].text)
    except:
        card.append(np.nan)
    try:
        os.append(x[7].text)
    except:
        os.append(np.nan)

# Create DataFrame
mobiles = pd.DataFrame({
    'Name': name,
    'Price': price,
    'Score': scores,
    'Rating': ratings,
    'SIM': sim,
    'Processor': processor,
    'RAM': ram,
    'Battery': battery,
    'Display': display,
    'Camera': camera,
    'Card': card,
    'OS': os
})

# View the DataFrame
mobiles
```

### Handling NAN in CARD and OS columns

```python
mobiles.isna().sum()
```

```python
mobiles['Card'] = mobiles.Card.fillna('Memory Card Not Supported')
```

```python
mobiles['OS'] = mobiles.OS.fillna(mobiles['Name'].str.split(" ").str[0])
```

# extract data of tablets

```python
with open('smartprix_tablets.html','r',encoding = 'utf8') as f:
    html = f.read()
    
soup = BeautifulSoup(html,'lxml')
```

```python
container1 = soup.find_all('div', {'class': 'sm-product has-tag has-features has-actions'})
container2 = soup.find_all('div',{'class':'sm-product has-tag oos has-features has-actions'})
```

```python
import pandas as pd
import numpy as np

# Initialize lists to store data
name = []        # List to store product names
price = []       # List to store product prices
scores = []      # List to store product scores
ratings = []     # List to store product ratings
sim = []         # List to store SIM details
processor = []  # List to store processor details
ram = []         # List to store RAM details
battery = []     # List to store battery details
display = []     # List to store display details
camera = []      # List to store camera details
card = []        # List to store memory card support details
os = []          # List to store operating system details

# Combine container1 and container2 into one list of containers
containers = [container1, container2]

# Iterate over each container
for container in containers:
    # Iterate over items within each container
    for item in container:
        try:
            # Extract product name
            name.append(item.find('h2').text)
        except AttributeError:
            name.append(np.nan)

        try:
            # Extract product price
            price.append(item.find('span', {'class': 'price'}).text)
        except AttributeError:
            price.append(np.nan)

        try:
            # Extract product score
            score_element_1 = item.find('div', {'class': 'score rank-1-bg'})
            score_element_2 = item.find('div', {'class': 'score rank-2-bg'})
            score_element_3 = item.find('div', {'class': 'score rank-4-bg'})
            score_element_4 = item.find('div', {'class': 'score rank-5-bg'})
            score_element_5 = item.find('div', {'class': 'score rank-3-bg'})

            if score_element_1:
                score = score_element_1.find('b').text
            elif score_element_2:
                score = score_element_2.find('b').text
            elif score_element_3:
                score = score_element_3.find('b').text
            elif score_element_4:
                score = score_element_4.find('b').text
            elif score_element_5:
                score = score_element_5.find('b').text
            else:
                score = np.nan
            scores.append(score)
        except AttributeError:
            scores.append(np.nan)

        try:
            # Extract product rating
            rating_span = item.find('span', class_='sm-rating')
            if rating_span:
                rating = rating_span['style'].split(':')[1].strip(';')
            else:
                rating = np.nan
            ratings.append(rating)
        except (AttributeError, IndexError, TypeError):
            ratings.append(np.nan)

        # Extract various product features
        features = item.find('ul', {'class': 'sm-feat specs'}).find_all('li')
        try:
            sim.append(features[0].text)         # SIM details
        except IndexError:
            sim.append(np.nan)
        try:
            processor.append(features[1].text)   # Processor details
        except IndexError:
            processor.append(np.nan)
        try:
            ram.append(features[2].text)         # RAM details
        except IndexError:
            ram.append(np.nan)
        try:
            battery.append(features[3].text)     # Battery details
        except IndexError:
            battery.append(np.nan)
        try:
            display.append(features[4].text)     # Display details
        except IndexError:
            display.append(np.nan)
        try:
            camera.append(features[5].text)      # Camera details
        except IndexError:
            camera.append(np.nan)
        try:
            card.append(features[6].text)        # Memory card support details
        except IndexError:
            card.append(np.nan)
        try:
            os.append(features[7].text)          # Operating system details
        except IndexError:
            os.append(np.nan)

# Create DataFrame
tablets = pd.DataFrame({
    'Name': name,
    'Price': price,
    'Score': scores,
    'Rating': ratings,
    'SIM': sim,
    'Network': processor,
    'Processor': ram,
    'RAM': battery,
    'Battery': display,
    'Display': camera,
    'Camera': card,
    'Card': os
})

tablets
```
